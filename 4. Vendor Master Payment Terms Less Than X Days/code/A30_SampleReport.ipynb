```python
# Import necessary modules
import pandas as pd
import numpy as np
import os
```

```python
# Open the LFA1 dataset to dataframe
LFA1 = pd.read_csv(r'./../../src/LFA1_AltColTitles.csv',low_memory=False)
```

```python
# Open the LFB1 dataset to dataframe
LFB1 = pd.read_csv(r'./../../src/LFB1_AltColTitles.csv',low_memory=False)
```

```python
# Open the T052 dataset to dataframe
T052=pd.read_csv(r'./../../src/T052_AltColTitles.csv',encoding='Latin1')
```

```python
# Join LFA1 to LFB1
LFA1_LFB1 = pd.merge( LFA1,
                      LFB1,
                      left_on='Account_Number_of_Vendor_or_Creditor',
                      right_on='Account_Number_of_Vendor_or_Creditor',
                      how='inner'
                    )
```

```python
# Join T052 to prior result
joinDF = pd.merge( T052,
                           LFA1_LFB1,
                           left_on='Terms_of_Payment_Key',
                           right_on='Terms_of_Payment_Key',
                           how='inner')
```

```python
# Handle the Central Deletion Flag to avoid duplicates
joinDF = joinDF.loc[joinDF['Central_Deletion_Flag_for_Master_Record'] != 'X',:]
joinDF.head(10)
```

```python
#Extract vendor less than 45 days as an example
lessThan45Days = joinDF[['Account_Number_of_Vendor_or_Creditor','Company_Code','Days_from_Baseline_Date_for_Payment']]
lessThan45Days = lessThan45Days[lessThan45Days['Days_from_Baseline_Date_for_Payment'] < 45]
```

```python
# Extract Data to a CSV File
lessThan45Days.to_csv(r'./../output/A30_SampleReport.csv', index=False)
```
